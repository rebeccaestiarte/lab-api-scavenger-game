## Challenge 3: Hidden Cold Joke

Using Python, call Github API to find out the cold joke contained in the 24 secret files in the following repo:

https://github.com/ironhack-datalabs/scavenger

The filenames of the secret files contain .scavengerhunt and they are scattered in different directories of this repo. The secret files are named from .0001.scavengerhunt to .0024.scavengerhunt. They are scattered randomly throughout this repo. You need to search for these files by calling the Github API, not searching the local files on your computer.

#### 1. Find the secret files.

`````python
user = ("rebeccaestiarte", "6ec9a3363fb411176d13d5774cdc59b41f546aa9")
response = requests.get('https://api.github.com/repos/Ironhack-datalabs/scavenger/contents', auth = user)
response
`````

`````python
result = response.json()
`````

`````python
result[0]
`````

`````python
names = []
for i in range(len(result)):
    names.append(result[i]["name"])
names    
`````

`````python
user = ("rebeccaestiarte", "6ec9a3363fb411176d13d5774cdc59b41f546aa9")
folder_content_list= []


for folder in names[1:]:
    url = f"https://api.github.com/repos/Ironhack-datalabs/scavenger/contents/{folder}"
    response = requests.get(url, auth = user)
    result = response.json()
    folder_content_list.append(result)

folder_content_list
`````

`````python
names = []
for i in range (len(folder_content_list)):
    for j in range (len(folder_content_list[i])):
        names_sc = folder_content_list[i][j]["name"]
        if names_sc.endswith("scavengerhunt"):
            names.append(folder_content_list[i][j]["name"])

names
`````

`````python
paths = []
for i in range (len(folder_content_list)):
    for j in range (len(folder_content_list[i])):
        paths_sc = folder_content_list[i][j]["name"]
        if paths_sc.endswith("scavengerhunt"):
            paths.append(folder_content_list[i][j]["path"])

paths
`````

`````python
test = dict(zip(names,paths))
test
`````

#### 2.  Sort the filenames ascendingly.

`````python
sorted_keys = sorted(test.keys())

sorted_keys
`````

#### 3. Read the content of each secret files into an array of strings.
Since the response is encoded, you will need to send the following information in the header of your request:
````python
headers = {'Accept': 'application/vnd.github.v3.raw'}
````

`````python
paths = []
for key in sorted_keys:
    values = test[key]
    paths.append(values)
paths 
`````

#### 4. Concatenate the strings in the array separating each two with a whitespace.

#### 5. Print out the joke.

`````python
for path in paths:
    url = f"https://api.github.com/repos/Ironhack-datalabs/scavenger/contents/{path}"
    response = requests.get(url, auth = user, headers = {'Accept': 'application/vnd.github.v3.raw'})
    result = response.text
    print (result)
`````
